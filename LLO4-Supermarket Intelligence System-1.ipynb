```python
from sklearn.cluster import KMeans
import pandas as pd

data = {
    "Spending": [100,200,1500,1800,120,1700,90,1600]
}

df = pd.DataFrame(data)

model = KMeans(n_clusters=2)
df["Group"] = model.fit_predict(df)

print(df)
```

```python
# Explain (Khmer)
# AI បែងចែកជា:
# •	Low Spender
# •	High Spender
# Business Use:
# → ផ្ញើ Promotion ខុសគ្នា
# PART B — Association Rule Mining (Simple Logic Version)
```

```python
transactions = [
    ["Milk", "Bread"],
    ["Milk", "Butter"],
    ["Bread", "Butter"],
    ["Milk", "Bread", "Butter"]
]

from collections import Counter
pair_count = Counter()
for transaction in transactions:
    for item in transaction:
        for other in transaction:
            if item != other:
                pair_count[(item, other)] += 1

print(pair_count)
```

```python
# Explain
# AI រាប់ថា Item មួយកើតជាមួយ Item មួយប៉ុន្មានដង
# Example Output:
# Milk → Bread (frequent)
# Business Use:
# → ដាក់ Product ជិតគ្នា
# → Bundle Promotion
# PART C — Data Mining Analysis
# Combine:
# 1️-Find customer groups (Clustering)
# 2️-Find product relationship (Association)
# 3️-Analyze trend (Pattern discovery)
# How Intelligence is Extracted
# Large Raw Data
# ↓
# Pattern Discovery
# ↓
# Grouping
# ↓
# Relationship Finding
# ↓
# Business Insight
```
