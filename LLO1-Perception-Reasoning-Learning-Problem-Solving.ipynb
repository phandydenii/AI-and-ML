```python
# Python Example Code
# Step 1: Install Library
# 	pip install scikit-learn pandas
# Step 2: Create Dataset & Train Model
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

# 1️-Perception: AI ទទួលទិន្នន័យ
data = {
    "StudyHours": [1, 2, 3, 4, 5, 6, 7, 8],
    "Attendance": [50, 60, 70, 75, 80, 85, 90, 95],
    "Result": ["Fail", "Fail", "Fail", "Pass", "Pass", "Pass", "Pass", "Pass"]
}
df = pd.DataFrame(data)
# Convert Result to numeric
df["Result"] = df["Result"].map({"Fail": 0, "Pass": 1})

X = df[["StudyHours", "Attendance"]]
y = df["Result"]

# 2️-Learning: Train Model
model = DecisionTreeClassifier()
model.fit(X, y)

# 3️-Reasoning & Problem Solving
new_student = [[3, 65]]
prediction = model.predict(new_student)

if prediction[0] == 1:
    print("Student will PASS")
else:
    print("Student will FAIL - Need Improvement")
# Code Explanation (Explain as Khmer)
# 🔹 Perception
# data = {...}
# AI ទទួលព័ត៌មាន (Study Hours, Attendance)
# 🔹 Learning
# model.fit(X, y)
# AI រៀនពី Data ដើម្បីរក Pattern
# 🔹 Reasoning
# model.predict(new_student)
# AI ប្រើ Pattern ដើម្បីទស្សន៍ទាយ
# 🔹 Problem Solving
# print("Need Improvement")
# AI ផ្តល់ដំណោះស្រាយ
# 📊 Output Example
# Student will FAIL - Need Improvement
    
# 1️⃣ DecisionTreeClassifier គឺអ្វី?
# Decision Tree គឺជា Tree-Based Model 🌳
# វាអាច បង្កើត Rule ដោយស្វ័យប្រវត្តិ ដូចជា៖
# StudyHours ≤ 3 → Fail
# StudyHours > 3 និង Attendance > 70 → Pass
# វាផ្ដល់ ការយល់ដល់ Rule បានច្បាស់
# ល្អសម្រាប់ Binary Classification (Pass/Fail) និង Multi-Class
# 2️⃣ ហេតុអ្វីមិនប្រើ Linear Regression?
# Linear Regression ផ្តោតលើ Predict Continuous Number
# ឧទាហរណ៍៖ predict score 0–100
# តែ Project នេះ Output = Pass / Fail (Binary) ❌
# ប្រើ Linear Regression → អាច predict 0.3, 1.2… ដែលមិនមានអត្ថន័យ
# 3️⃣ ហេតុអ្វីមិនប្រើ Logistic Regression?
# Logistic Regression អាច predict Binary ✅ (0/1)
# ល្អសម្រាប់ linear relationship (Features និង Probability មាន linear correlation)
# តែ Project សិស្សនេះ Decision Boundary មិន linear ប៉ុន្មាន
# Example:
# StudyHours ≤ 3 → Fail
# StudyHours > 3, Attendance ≤ 70 → Fail
# StudyHours > 3, Attendance > 70 → Pass
# Logistic Regression មិនអាច capture splitting rules ច្រើនជំហាន ងាយស្រួល, ត្រូវ Decision Tree
```

```python
# 1. ដំឡើងបណ្ណាល័យត្រូវការ
# pip install scikit-learn pandas numpy

# 2. បង្កើត Data តូចៗ (គំរូ Spam/Ham)
# (ប្រើ dataset តូចៗ ដើម្បីងាយស្រួលសិក្សា)

import pandas as pd
data = {
    'text': [
        "Win money now!!! Click here",
        "Free offer available, claim your prize",
        "Hello teacher, I already submitted my homework",
        "Meeting tomorrow at 9 AM",
        "Congratulations! You won a free vacation",
        "Can we discuss the AI project today?"
    ],
    'label': ["spam", "spam", "ham", "ham", "spam", "ham"]
}
df = pd.DataFrame(data)
df
# 3.បំបែកទិន្នន័យ Train/Test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
df['text'], df['label'], test_size=0.3, random_state=42
)

# 4. បំលែងអត្ថបទទៅជាលក្ខណៈ (Features)
# ប្រើ TF-IDF Vectorizer

from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# 5. បណ្តុះម៉ូដែល (Model Training)
# ប្រើ Logistic Regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train_vec, y_train)

# 6. វាយតម្លៃម៉ូដែល
from sklearn.metrics import classification_report
y_pred = model.predict(X_test_vec)
print(classification_report(y_test, y_pred))
sample = ["Click here to claim your free reward"]
sample_vec = vectorizer.transform(sample)
print("Prediction:", model.predict(sample_vec)[0])
```

```python
# 🧠 Step 1: Import Library
# # pip install scikit-learn pandas numpy matplotlib

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# 📊 Step 2: បង្កើត Dataset (គំរូសិស្ស)
data = {
    "study_hours": [1,2,3,4,5,6,7,8,2,3,5,6,7,1,4],
    "attendance": [50,60,65,70,80,85,90,95,55,60,75,80,85,40,72],
    "homework_score": [40,50,55,60,70,75,80,90,45,50,65,70,85,35,68],
    "result": ["Fail","Fail","Fail","Pass","Pass","Pass","Pass","Pass",
               "Fail","Fail","Pass","Pass","Pass","Fail","Pass"]
}

df = pd.DataFrame(data)
df

# 📌 result = Target (Output)
# Pass / Fail
```

```python
# ✂ Step 3: បំបែក Train / Test
from sklearn.model_selection import train_test_split

X = df[["study_hours","attendance","homework_score"]]
y = df["result"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# 👉 70% សម្រាប់បង្រៀន
# 👉 30% សម្រាប់ប្រឡង
# 🤖 Step 4: Train Model (Logistic Regression)
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)

# 📈 Step 5: វាយតម្លៃម៉ូដែល
from sklearn.metrics import classification_report, accuracy_score

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# 📌 Accuracy ជាពិន្ទុ AI

# 🔮 Step 6: សាកល្បងសិស្សថ្មី
new_student = [[4, 75, 65]]  # 4h study, 75% attendance, score 65
prediction = model.predict(new_student)

print("Prediction:", prediction[0])

# 👉 Output អាចជា:

# Prediction: Pass
```

```python
# 🎓 Project: AI ទស្សន៍ទាយ Pass / Fail ពី Essay សិស្ស

# 👉 Input = អត្ថបទ Essay
# 👉 Output = Pass / Fail
# 👉 នេះជាបញ្ហា Text Classification

# 🧠 Step 1: Install & Import Library
# pip install scikit-learn pandas

import pandas as pd

# 📚 Step 2: បង្កើត Dataset Essay
data = {
    "essay": [
        "I studied very hard and understood machine learning concepts clearly",
        "I did not study much and I am confused about the lessons",
        "I completed all assignments and practiced coding daily",
        "I skipped many classes and did not submit homework",
        "I reviewed past exams and practiced problem solving",
        "I did not prepare for the exam at all"
    ],
    "result": ["Pass", "Fail", "Pass", "Fail", "Pass", "Fail"]
}

df = pd.DataFrame(data)
df
# ✂ Step 3: Train/Test Split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df["essay"], df["result"], test_size=0.3, random_state=42
)
# 🔤 Step 4: ប្រើ TF-IDF Vectorizer

# 👉 ព្រោះ data ជា “Text”
# Computer មិនយល់អក្សរ ❌
# ត្រូវបំលែងទៅជា Number ✔

from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# 📌 fit_transform() → រៀនពាក្យ
# 📌 transform() → បំលែង Test data

# 🤖 Step 5: Train Logistic Regression
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train_vec, y_train)
# 📊 Step 6: វាយតម្លៃម៉ូដែល
from sklearn.metrics import classification_report, accuracy_score

y_pred = model.predict(X_test_vec)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# 🔮 Step 7: សាកល្បង Essay ថ្មី
new_essay = ["I practiced a lot and studied every day before exam"]

new_vec = vectorizer.transform(new_essay)
prediction = model.predict(new_vec)

print("Prediction:", prediction[0])

# 👉 Output អាចជា:

# Prediction: Pass
```
