```python
# 🎬 Full AI Project
# Movie Recommendation System (Content-Based Filtering)
# 1️⃣ Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python

# ពន្យល់

# Library	Function
# pandas	គ្រប់គ្រង dataset
# numpy	គណនា
# matplotlib	visualization
# CountVectorizer	បំលែង text → number
# cosine_similarity	គណនា similarity
# 2️⃣ Create Movie Dataset
movies = {

"title":[
    "Avengers",
    "Batman",
    "Superman",
    "SpiderMan",
    "IronMan",
    "Titanic",
    "Notebook",
    "La La Land",
    "Fast Furious",
    "John Wick"
    ],

    "genre":[
        "Action Hero Adventure",
        "Action Hero Dark",
        "Action Hero Alien",
        "Action Hero Teen",
        "Action Hero Tech",
        "Romance Drama",
        "Romance Love",
        "Romance Music",
        "Action Car Racing",
        "Action Gun Fight"
    ]
}
df = pd.DataFrame(movies)
df
```

```python
# Dataset Example
# Movie	Genre
# Avengers	Action Hero Adventure
# Batman	Action Hero Dark
# Titanic	Romance Drama
# 3️⃣ Data Exploration
df.info()
df.describe()

# គោលបំណង

# ពិនិត្យ dataset

# ឃើញ structure data

# 4️⃣ Feature Engineering

# យើងបំលែង genre text → vector

vectorizer = CountVectorizer()

matrix = vectorizer.fit_transform(df["genre"])

matrix

# Example

# Movie	Action	Hero	Romance
# Avengers	1	1	0
# Titanic	0	0	1
```

```python
# 5️⃣ Calculate Movie Similarity
similarity = cosine_similarity(matrix)

similarity

# Similarity score

# 0 = not similar
# 1 = exactly same
# 6️⃣ Create Recommendation Function
def recommend(movie):

    index = df[df["title"] == movie].index[0]

    distances = list(enumerate(similarity[index]))

    distances = sorted(distances, key=lambda x:x[1], reverse=True)

    recommendations = distances[1:6]

    for i in recommendations:
        print(df.iloc[i[0]].title)
```

```python
# 7️⃣ Test Recommendation
recommend("Avengers")

# Output

# IronMan
# SpiderMan
# Batman
# John Wick
# Fast Furious

# មានន័យថា

# System ណែនាំ Action Movies

# 8️⃣ Another Test
recommend("Titanic")

# Output

# Notebook
# La La Land

# ព្រោះ

# Titanic = Romance Movie
```
