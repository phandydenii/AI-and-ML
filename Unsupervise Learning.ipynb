```python
Unsupervised Learning គឺជាប្រភេទ Machine Learning ដែល ទិន្នន័យមិនមាន label ឬ output ដែលច្បាស់។ គោលដៅរបស់វាគឺ រក pattern, cluster, structure 
ឬ association នៅក្នុងទិន្នន័យដោយឯករាជ្យ។
នៅក្នុង Unsupervised Learning មានម៉ូដែល (models) មួយចំនួនសំខាន់ៗដែលយើងអាចចែកបានជាក្រុមៗ៖
1️⃣ Clustering Models (ម៉ូដែលក្រុមបែងចែក)
បែងចែកទិន្នន័យទៅក្រុម (clusters) ដោយផ្អែកលើ similarity ឬ distance របស់ទិន្នន័យ
K-Means Clustering → បែងចែកទិន្នន័យជាក្រុម k cluster
Hierarchical Clustering → បង្កើត tree-like structure របស់ clusters
DBSCAN (Density-Based Spatial Clustering of Applications with Noise) → cluster based on density, can detect outliers
2️⃣ Dimensionality Reduction Models (ម៉ូដែលកាត់បន្ថយ dimension)
កាត់បន្ថយ features ដើម្បី visual ឬស្រួល process
PCA (Principal Component Analysis) → find main components that explain most variance
t-SNE (t-Distributed Stochastic Neighbor Embedding) → visualize high-dimensional data in 2D/3D
UMAP (Uniform Manifold Approximation and Projection) → similar to t-SNE but faster
3️⃣ Association Rule Learning (ម៉ូដែលរក pattern association)
រក relation ឬ pattern ដែលកើតឡើងជាមួយ items ក្នុង dataset
Apriori Algorithm → find frequent itemsets & association rules
Eclat Algorithm → alternative for frequent itemset mining
4️⃣ Anomaly Detection Models (ម៉ូដែលរកអចិន្រ្តៃ/ពិសេស)
រកទិន្នន័យដែល ខុសពី pattern ធម្មតា
Isolation Forest
One-Class SVM
```

```python
# 1️⃣ Clustering (ការបែងចែកក្រុម)
# (a) K-Means Clustering)
# គោលបំណង: បែងចែកទិន្នន័យទៅក្រុម k based on similarity.
# ប្រើប្រាស់: Customer segmentation, image compression, document clustering.

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

# Sample 2D data
X = np.array([[1,2],[1,4],[1,0],[10,2],[10,4],[10,0]])

# KMeans clustering
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# Visualization
plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis', s=100)
plt.scatter(centers[:,0], centers[:,1], c='red', marker='X', s=200)
plt.title("K-Means Clustering")
plt.show()
```

```python
# (b) DBSCAN (Density-Based Clustering)
# គោលបំណង: Cluster data based on density, detect outliers.
# ប្រើប្រាស់: Detect anomalies, geospatial data clustering.
from sklearn.cluster import DBSCAN

X = np.array([[1,2],[2,2],[2,3],[8,7],[8,8],[25,80]])
dbscan = DBSCAN(eps=3, min_samples=2)
dbscan.fit(X)
labels = dbscan.labels_

plt.scatter(X[:,0], X[:,1], c=labels, cmap='rainbow', s=100)
plt.title("DBSCAN Clustering (Red=-1 noise)")
plt.show()
```

```python
# 2️⃣ Dimensionality Reduction (កាត់បន្ថយ dimension)
# (a) PCA (Principal Component Analysis)
# គោលបំណង: Reduce features while keeping variance.
# ប្រើប្រាស់: Visualization, noise reduction.
    
from sklearn.decomposition import PCA
import pandas as pd

# Sample 3D data
X = np.array([[2.5, 2.4, 0.5],
              [0.5, 0.7, 2.2],
              [2.2, 2.9, 1.5],
              [1.9, 2.2, 0.3],
              [3.1, 3.0, 0.7]])

pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

plt.scatter(X_reduced[:,0], X_reduced[:,1], s=100)
plt.title("PCA Dimensionality Reduction (3D to 2D)")
plt.show()
```

```python
# (b) t-SNE (Visualization high-dimensional data)
from sklearn.manifold import TSNE

X = np.random.rand(10,5)  # 10 samples, 5 features
tsne = TSNE(n_components=2, random_state=42)
X_embedded = tsne.fit_transform(X)

plt.scatter(X_embedded[:,0], X_embedded[:,1], c='green', s=100)
plt.title("t-SNE 2D Visualization")
plt.show()
```

```python
# 3️⃣ Association Rule Learning (រក pattern association)

# Apriori Algorithm (using mlxtend library)

# គោលបំណង: Find frequent itemsets & association rules

# ប្រើប្រាស់: Market basket analysis
# !pip install mlxtend

from mlxtend.frequent_patterns import apriori, association_rules
import pandas as pd
import seaborn as sns

data = pd.DataFrame({
    'Milk':[1,0,1,1],
    'Bread':[1,1,1,0],
    'Eggs':[0,1,1,1]
})

freq_itemsets = apriori(data, min_support=0.5, use_colnames=True)
rules = association_rules(freq_itemsets, metric="lift", min_threshold=1.0)

# Visualize lift vs confidence
sns.scatterplot(data=rules, x='support', y='confidence', size='lift', hue='lift', palette='viridis', sizes=(100,500))
plt.title("Association Rules: Support vs Confidence (size=Lift)")
plt.show()
```

```python
# 4️⃣ Anomaly Detection (រកទិន្នន័យពិសេស / outliers)

# (a) Isolation Forest)

# គោលបំណង: Detect anomalies automatically

# ប្រើប្រាស់: Fraud detection, unusual event detection
from sklearn.ensemble import IsolationForest
import numpy as np

X = np.array([[10],[12],[10],[100]])  # 100 is anomaly
clf = IsolationForest(contamination=0.1, random_state=42)
clf.fit(X)
pred = clf.predict(X)  # 1 = normal, -1 = anomaly

colors = ['red' if p==-1 else 'blue' for p in pred]
plt.scatter(X, np.zeros_like(X), c=colors, s=200)
plt.title("Isolation Forest Anomaly Detection (Red=Anomaly)")
plt.show()
```

```python
# (b) One-Class SVM)

# គោលបំណង: Detect outliers based on support vector machine

# ប្រើប្រាស់: Similar to Isolation Forest
from sklearn.svm import OneClassSVM

X = np.array([[1],[2],[2],[3],[10]])  # 10 is anomaly
ocsvm = OneClassSVM(nu=0.1, kernel="rbf", gamma=0.1)
ocsvm.fit(X)
pred = ocsvm.predict(X)

colors = ['red' if p==-1 else 'green' for p in pred]
plt.scatter(X, np.zeros_like(X), c=colors, s=200)
plt.title("One-Class SVM Anomaly Detection (Red=Anomaly)")
plt.show()
```
