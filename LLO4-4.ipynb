```python
# 2️⃣ Import Libraries
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
# ពន្យល់
# Library	Function
# pandas	គ្រប់គ្រង dataset
# numpy	គណនា matrix
# cosine_similarity	គណនា similarity
```

```python
# 3️⃣ Create Rating Dataset
ratings = {

"user":[1,1,1,2,2,3,3,3,4,4],
    "movie":[
        "Avengers",
        "Batman",
        "IronMan",
        "Avengers",
        "Superman",
        "Batman",
        "Superman",
        "SpiderMan",
        "IronMan",
        "SpiderMan"
    ],
    "rating":[5,4,5,4,5,5,4,5,4,5]
}

df = pd.DataFrame(ratings)

df
# Dataset
# User	Movie	Rating
# 1	Avengers	5
# 1	Batman	4
# 1	IronMan	5
# 2	Avengers	4
# 2	Superman	5
```

```python
# 4️⃣ Create User-Movie Matrix
movie_matrix = df.pivot_table(
    index="user",
    columns="movie",
    values="rating"
)

movie_matrix
# Matrix Example
# User	Avengers	Batman	IronMan	SpiderMan
# 1	5	4	5	NaN
# 2	4	NaN	NaN	NaN
# 3	NaN	5	NaN	5
```

```python
# នេះហៅថា User-Item Matrix
# 5️⃣ Replace Missing Values
movie_matrix = movie_matrix.fillna(0)
# 6️⃣ Calculate User Similarity
similarity = cosine_similarity(movie_matrix)

similarity_df = pd.DataFrame(
    similarity,
    index=movie_matrix.index,
    columns=movie_matrix.index
)

similarity_df
# Output
# User	1	2	3	4
# 1	1	0.8	0.5	0.6
# មានន័យ
# User 1 និង User 2
# មាន similarity 0.8
```

```python
# 7️⃣ Find Similar Users
# similarity_df[1].sort_values(ascending=False)
# Result
# 1
# 2
# 4
# 3
# User 2 គឺ similar user
# 8️⃣ Recommendation Algorithm
def recommend_movies(user_id):
    similar_users = similarity_df[user_id].sort_values(ascending=False)
    similar_users = similar_users.index[1:3]
    movies = movie_matrix.loc[similar_users]
    recommendation = movies.mean().sort_values(ascending=False)
    return recommendation
```

```python
# 9️⃣ Test Recommendation
recommend_movies(1)
# Output
# SpiderMan
# Superman
# មានន័យថា
# System ណែនាំ
# SpiderMan
# Superman
# ព្រោះ Similar Users ចូលចិត្ត Movie នេះ

# 🔟 Data Mining Techniques Used
# Technique	Purpose
# Collaborative Filtering	Recommend movies
# Similarity Analysis	Find similar users
# Pattern Discovery	Find rating patterns

# 📊 System Flow
# User Rating Data
#        ↓
# User-Movie Matrix
#        ↓
# Similarity Calculation
#        ↓
# Find Similar Users
#        ↓
# Recommend Movies
# 🎯 Real-World Applications
# Recommendation System ប្រើក្នុង
# Platform	Purpose
# Netflix	Movie recommendation
# YouTube	Video recommendation
# Amazon	Product recommendation
# Spotify	Music recommendation
# 🧠 Knowledge Discovery
# Data Mining រកឃើញ
# ✔️ User preference pattern
# ✔️ Similar users
# ✔️ Personalized recommendations
```
